## Stack plots

One method we've come up with to visualise the GGDOT data is something we call a stack plot. Here, we'll show you an example of how to construct one, then walk you through building one yourself.

```python
# make the plot nice and big
plt.rcParams['figure.figsize'] = [14,5]

# here's an example of how to make a stack plot
stack_plot (["CO2e", "Beef Burger", "kCals"],
            [[8, "Lettuce, 20 g", 2],
             [15, "1 Tomato Slice", 4],
             [19, "2 tbsp Ketchup", 17],
             [2875, "Beef Patty, 125 g", 221],
             [115, "Bacon, 25g", 75],
             [262, "Cheddar Cheese, 30 g", 125],
             [90, "Bun, 120 g", 360]]
            ,scale=[1, 0.5, 0.23]
           )

# If you want to save your plot, uncomment the next line
#plt.savefig("burgerstack.svg")
```

The stack plots help us visualise the CO2e emissions of food relative to how many calories they have.
